### Selecione os imoveis classificados com tipo "apartamento"
### Selecione os imoveis classificados com tipos 'casa','casa de condominio' e 'casa de vila'
### Selecione os imoveis com areas entre 60 e 100  metros quadradaso , incluindo os limites 
### Selecone os moveis que tenha pelo menos 4 quartos e aluguel menor que 2.000

```python
dados = pd.read_csv("dados/aluguel_residencial.csv",sep=';')
dados
```

```python
dados['Tipo'].unique()
```

```python
#Selecione os imoveis classificados com tipo "apartamento"
selecao = dados['Tipo'] == 'Apartamento'
n1 = dados[selecao].shape[0]
n1
```

```python
#Selecione os imoveis classificados com tipos 'casa','casa de condominio' e 'casa de vila'
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao].shape[0]
n2
```

```python
#Selecione os imoveis com areas entre 60 e 100 metros quadradaso , incluindo os limites
selecao = (dados['Area']>=60) & (dados['Area']<=100)
n3 = dados[selecao].shape[0]
n3
```

```python
#Selecone os moveis que tenha pelo menos 4 quartos e aluguel menor que 2.000
selecao = (dados['Quartos']>=4) & (dados['Valor']<2000)
n4 = dados[selecao].shape[0]
n4
```

```python
print(f"Nº imoveis classificados com tipo 'apartamento'  : {n1}")
print(f"Nº imoveis classificados com tipos 'casa','casa de condominio' e 'casa de vila' :  {n2}")
print(f"Nº imoveis com areas entre 60 e 100 metros quadradaso , incluindo os limites :  {n3}")
print(f"Nº moveis que tenha pelo menos 4 quartos e aluguel menor que 2.000 : {n4}")
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados['Tipo'] == tipo
    eh_dentro_limite = (dados['Valor'] >= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```
