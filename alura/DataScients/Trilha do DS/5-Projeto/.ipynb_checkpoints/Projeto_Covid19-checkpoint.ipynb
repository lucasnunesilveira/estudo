```python
#Quando a escalar estar em 1e7 ?
confirmed_by_country.loc['China'][760:800].plot()
plt.xlabel("Datas dos casos", fontsize= 16)
plt.ylabel("Quantidade de dados", fontsize= 16)
plt.ticklabel_format(style='plain', axis='y')
plt.grid(True)
```

```python
#Quando a escalar estar em 1e7 ?
plt.figure(figsize=(15,5))
confirmed_by_country.loc['Brazil'][700:800].plot()
plt.xlabel("Datas dos casos", fontsize= 16)
plt.ylabel("Quantidade de dados", fontsize= 16)
plt.ticklabel_format(style='plain', axis='y')
plt.grid(True)
```

```python
#https://www.worldometers.info/coronavirus/
plt.figure(figsize=(15,5))
new_cases_Brazil= confirmed_by_country.loc['Brazil'][700:800].diff().dropna()
plt.title("Casos no Brazil do 21/12/21 a 11/3/22",fontsize=16)
plt.xlabel("Datas dos casos", fontsize= 16)
plt.ylabel("Quantidade de dados", fontsize= 16)
plt.ticklabel_format(style='plain', axis='y')
new_cases_Brazil.plot()
plt.show()
     
```

```python
plt.figure(figsize=(10,5))
last_date = "6/5/21"
confirmed_by_country[last_date].sort_values(ascending=False)[1:10].plot(kind='bar')
plt.title("Confirmados por país")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
